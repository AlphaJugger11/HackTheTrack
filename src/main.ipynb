```python
# Load the telemetry data

telemR1 = pd.read_csv('../dataset/barber/R1_barber_telemetry_data.csv')
telemR2 = pd.read_csv('../dataset/barber/R2_barber_telemetry_data.csv')

# Check unique drivers in R1
print("=== Race 1 Drivers ===")
drivers_r1 = telemR1['Driver'].unique()
print(f"Total drivers: {len(drivers_r1)}")
print(f"Drivers: {sorted(drivers_r1)}\n")

# Check telemetry availability by driver and lap for R1
print("=== Race 1 - Telemetry Data by Driver and Lap ===")
r1_summary = telemR1.groupby(['Driver', 'Lap']).size().reset_index(name='data_points')
print(r1_summary.to_string(index=False))
print(f"\nTotal laps with telemetry in R1: {r1_summary.shape[0]}\n")

# Check unique drivers in R2
print("=== Race 2 Drivers ===")
drivers_r2 = telemR2['Driver'].unique()
print(f"Total drivers: {len(drivers_r2)}")
print(f"Drivers: {sorted(drivers_r2)}\n")

# Check telemetry availability by driver and lap for R2
print("=== Race 2 - Telemetry Data by Driver and Lap ===")
r2_summary = telemR2.groupby(['Driver', 'Lap']).size().reset_index(name='data_points')
print(r2_summary.to_string(index=False))
print(f"\nTotal laps with telemetry in R2: {r2_summary.shape[0]}\n")

# Optional: Show lap range for each driver
print("=== Lap Range by Driver (R1) ===")
r1_lap_range = telemR1.groupby('Driver')['Lap'].agg(['min', 'max', 'count']).reset_index()
r1_lap_range.columns = ['Driver', 'First_Lap', 'Last_Lap', 'Total_Data_Points']
print(r1_lap_range.to_string(index=False))

print("\n=== Lap Range by Driver (R2) ===")
r2_lap_range = telemR2.groupby('Driver')['Lap'].agg(['min', 'max', 'count']).reset_index()
r2_lap_range.columns = ['Driver', 'First_Lap', 'Last_Lap', 'Total_Data_Points']
print(r2_lap_range.to_string(index=False))
```
